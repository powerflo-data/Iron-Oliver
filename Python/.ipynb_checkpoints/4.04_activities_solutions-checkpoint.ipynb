```python
# Standardization/Standard Scaler
X_num = X_num.drop(['AVGGIFT'], axis=1)
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(X_num)
x_standardized = transformer.transform(X_num)
pd.DataFrame(x_standardized)
```

```python
def standard_scaler(x, u, s):
    return (x- u)/s
```

```python
X_new = pd.DataFrame(columns=X_num.columns)
```

```python
X_new
```

```python
X_num.head()
# X_num.columns
```

```python
for col in X_num:
    X_num['mean'] = np.mean(X_num[col])
    X_num['std']  = np.std(X_num[col])
    X_new[col] = list(map(standard_scaler, X_num[col], X_num['mean'], X_num['std']))
    X_num = X_num.drop(['mean', 'std'], axis=1)
```
