```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import binom
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
```

```python
n = [10,100,1000,10000]
```

```python
random = np.random.binomial(n=100, p=0.5, size=1)
random
```

```python
#fig, ax = plt.subplots(4)
for i,val in enumerate(n):
    random = np.random.binomial(n=100, p=0.5, size=val)
    sns.histplot(random,kde = True)
    plt.show()
    print('Mean: ',np.mean(random))
```
