# In Python: Decision Tree for classification

```python
import pandas as pd
from sklearn.datasets import load_iris

data = load_iris()
data['DESCR']

X, y = load_iris(return_X_y = True)
```

```python
df_X = pd.DataFrame(data=X, columns=data['feature_names'])
```

```python
df_y = pd.DataFrame(data=y, columns=['species'])
```

```python
df_X
```

```python
df_y.value_counts()
```

```python
Image(filename='flower-part-purpose-seeds-making-insects-.jpg', width=500) 
```

```python
data['target_names']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## let's plot the decision tree

```python
from sklearn.tree import plot_tree

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(4,4),
                      dpi=300,
                      )

fn = data['feature_names']
cn = data['target_names']

plot_tree(model,
         feature_names=fn,
         class_names=cn,
         fontsize=3,
          filled=True,
         ax=ax,
         );
```
