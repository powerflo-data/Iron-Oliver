## Respectful scraping!

Scraping wikipedia creates load on the wiki server, so, let's build in an artificial lag

````python
from time import sleep
for i in range(5):
    print(i)
    sleep(2)
````

````python
from random import randint

for i in range(5):
    print(i)
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " seconds")
    sleep(wait_time)
````

### let's make use of both random and sleep in our script!

````python
pages =  []   # in here go the response objects of all of our search result pages

for number in iterations:
    
    # assemble the url
    number = str(number)
    url = f"https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start={number}&ref_=adv_nxt"
    
    # download the html with a get request
    response = requests.get(url)
    
    # monitor the process by printing the status code
    print("Status code: " + str(response.status_code))
    
    # build the collection of responses aka the search results per page
    pages.append(response)
    
    # let's scrape respectfully
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " seconds")
    sleep(wait_time)
````

Another approach to append to that list, without making use of the `+` is, to check, whether the current `soup.select()` actually contains something. If so, append to the list.

````python
len_table = len(soup.select("tbody > tr"))
````

````python
presidents = []
for i in range(len_table):
    to_append = soup.select("tbody > tr:nth-child(" + str(i) + ") > td:nth-child(4) > b > a")
    if len(to_append) != 0:
        presidents.append(to_append)
````

````python
presi_soups = []

for president_url in presidents_urls:
    response = requests.get(president_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    presi_soups.append(soup.find("table", {'class' : 'infobox vcard'}))
    
    # respectful nap
    wait_time = randint(1,4)
    print("I will sleep for  " + str(wait_time) + " seconds.")
    sleep(wait_time)
````

````python
len(presi_soups)
````

````python
# finding out the birthday for only one president
presi_soups[-1].find('span', {'class':'bday'}).get_text()

# also the political party
presi_soups[-1].find('th', string="Political party").parent.a.get_text()

# also the number of children
len(presi_soups[-1].find('th', string='Children').parent.find_all('li'))
````

````python
birthdays = []
parties = []
children = []
presidents_names = []


for presi_soup in presi_soups:
    # finding out the birthday for only one president
    birthdays.append(presi_soup.find('span', {'class':'bday'}).get_text())

    # also the political party
    parties.append(presi_soup.find('th', string="Political party").parent.a.get_text())
    
    # president names
    presidents_names.append(presi_soup.find('div', {'class':'fn'}).get_text())
    # also the number of children
    
    # exception handling when a president has no children, then .find() throws an
    # AttributeError
    try:
        children.append(len(presi_soup.find('th', string='Children').parent.find_all('li')))
    except AttributeError:
        children.append(0)
````
