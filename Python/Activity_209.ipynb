```python
from sqlalchemy import create_engine
import pandas as pd
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)
```

```python
# Activity 1:
#1 & #2
result = engine.execute('SELECT * FROM bank.loan where status = \'A\' or status = \'B\'')
for row in result:
    print(row)
```

```python
# Activity 2:
#First query
data = pd.read_sql_query('select count(acc.account_id)  as Number_of_Accounts, year(acc.date) as Year, dis.A2 as District \
from bank.account as acc \
inner join bank.district as dis \
on acc.district_id = dis.A1 \
group by year(acc.date), dis.A2 \
order by Year, Number_of_Accounts DESC;', engine)
data.head()
```

```python
#second query
data = pd.read_sql_query('select sum(loa.amount) as Total_amount, round(avg(loa.amount),0) as Average_Amount, dis.A2 as "District", rank() over (order by sum(loa.amount) DESC) as "Rank" \
from bank.account as acc \
inner join bank.district as dis \
inner join bank.loan as loa \
on acc.account_id = loa.account_id and acc.district_id = dis.A1 \
group by dis.A2 \
order by Total_amount DESC;', engine)
data.head()
```
