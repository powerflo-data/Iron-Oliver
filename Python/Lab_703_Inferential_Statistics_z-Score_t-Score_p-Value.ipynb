You can also use scipy.stats.ttest_1samp(a, popmean, ...)
which performs a 1 sample t-test. popmean is the expected value in the null hypothesis and a is usually a set of values of your sample which is used to calculate the sample mean from. Since we don't have all the sample values, you can try to input just a list with the mean sample value: [130.1]

```python
my_pop = 120
n = 100
my_sample = 130.1
std_sample = 21.21
```

H0: The group does not differ from the regular population, therefore, the calculated sample mean is a result of a sample error

```python
statistic = (my_sample - my_pop) / (std_sample/math.sqrt(n))
statistic
```
