```python
def get_artists_from_song(songname):
    results = sp.search(q=songname, limit=50)
    return [artist['name'] for items in results['tracks']['items'] for artist in items['artists']]

def get_artists_id_song(songname):
    results = sp.search(q=songname, limit=50)
    return [artist['id'] for items in results['tracks']['items'] for artist in items['artists']]
```

## Get Tracks from a Playlist 

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    while results['next']:
        #next page:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
all_tracks = get_playlist_tracks('spotify', '4rnleEAOdmFAbRcNCgZMpY')
```
